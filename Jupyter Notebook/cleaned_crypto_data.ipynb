```python
import pandas as pd
import numpy as np
```

```python
# =========================
# LOAD DATASET
# =========================

df = pd.read_csv("C:/Users/lordb/Downloads/ethereum_2015-08-07_2024-06-28.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

```python
# =========================
# RENAME COLUMNS
# =========================

df.columns = [
    "start_date",
    "end_date",
    "open_price",
    "high_price",
    "low_price",
    "close_price",
    "volume",
    "market_cap"
]

print("\nColumns Renamed")
print(df.columns)
```

```python
# =========================
# CONVERT DATE COLUMNS
# =========================

df["start_date"] = pd.to_datetime(df["start_date"])
df["end_date"] = pd.to_datetime(df["end_date"])

print("\nDate Conversion Completed")
```

```python
# =========================
# CHECK MISSING VALUES
# =========================

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# =========================
# REMOVE DUPLICATES
# =========================

df.drop_duplicates(inplace=True)

print("\nDuplicates Removed")
```

```python
 # =========================
# CONVERT NUMERIC COLUMNS
# =========================

numeric_cols = [
    "open_price",
    "high_price",
    "low_price",
    "close_price",
    "volume",
    "market_cap"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("\nNumeric Conversion Completed")
```

```python
# =========================
# REMOVE NULL ROWS
# =========================

df.dropna(inplace=True)

print("\nNull Rows Removed")
```

```python
# =========================
# DAILY RETURN
# =========================

df["daily_return"] = (
    (df["close_price"] - df["open_price"])
    / df["open_price"]
) * 100

print("\nDaily Return Created")
```

```python
# =========================
# VOLATILITY
# =========================

df["volatility"] = (
    (df["high_price"] - df["low_price"])
    / df["open_price"]
) * 100

print("\nVolatility Column Created")
```

```python
# =========================
# BULLISH / BEARISH LABEL
# =========================

df["market_trend"] = np.where(
    df["close_price"] > df["open_price"],
    "Bullish",
    "Bearish"
)

print("\nMarket Trend Column Created")
```

```python
# =========================
# 7-DAY MOVING AVERAGE
# =========================

df["moving_avg_7"] = (
    df["close_price"]
    .rolling(window=7)
    .mean()
)

print("\nMoving Average Created")
```

```python
# =========================
# CRASH DETECTION
# =========================

df["crash_flag"] = np.where(
    df["daily_return"] < -10,
    "Crash",
    "Normal"
)

print("\nCrash Flag Created")
```

```python
# =========================
# MARKET CAP GROWTH
# =========================

df["market_cap_growth"] = (
    df["market_cap"].pct_change()
) * 100

print("\nMarket Cap Growth Created")
```

```python
# =========================
# FINAL DATASET INFO
# =========================

print("\nFinal Dataset Info")
print(df.info())

print("\nFirst 5 Rows")
print(df.head())
```

```python
# =========================
# SAVE CLEANED DATA
# =========================

output_path =  "C:/Users/lordb/Downloads/ethereum_2015-08-07_2024-06-28.csv"

df.to_csv(output_path, index=False)

print("\nCleaned Data Saved Successfully")
print(f"Saved at: {output_path}")
```
